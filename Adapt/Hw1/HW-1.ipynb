```python
def task1(year):
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        print(f"{year} - Високосный год")
    else:
        print(f"{year} - Обычный год")
```

```python
task1(2020)
task1(2022)
task1(2019)
task1(2000)
```

```python
def task2(ticket_number):
    ticket_str = str(ticket_number)

    if len(ticket_str) != 6:
        print(f"{ticket_number} - Неверный номер билета")
        return

    sum_first = int(ticket_str[0]) + int(ticket_str[1]) + int(ticket_str[2])
    sum_last = int(ticket_str[3]) + int(ticket_str[4]) + int(ticket_str[5])

    if sum_first == sum_last:
        print(f"{ticket_number} - Счастливый билет")
    else:
        print(f"{ticket_number} - Обычный билет")
```

```python
task2(123321)
task2(123456)
task2(111111)
task2(123)
```
