```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
```

```python
def get_owner_by_number(doc_number):
    for doc in documents:
        if doc['number'] == doc_number:
            return doc['name']
    return None
```

```python
def get_shelf_by_number(doc_number):
    for shelf, numbers in directories.items():
        if doc_number in numbers:
            return shelf
    return None
```

```python
def command_p():
    print("Введите номер документа:\n")
    doc_number = input()
    owner = get_owner_by_number(doc_number)
    if owner:
        print(f"Владелец документа: {owner}")
    else:
        print("Документ не найден")
```

```python
def command_s():
    doc_number = input("Введите номер документа:\n")
    shelf = get_shelf_by_number(doc_number)
    if shelf:
        print(f"Документ хранится на полке: {shelf}")
    else:
        print("Документ не найден на полках")
```

```python
def main():
    while True:
        command = input("Введите команду:\n")

        if command == 'q':
            print("Программа завершена")
            break
        elif command == 'p':
            command_p()
        elif command == 's':
            command_s()
        else:
            print("Неизвестная команда. Доступные команды: p, s, q")

main()
```
