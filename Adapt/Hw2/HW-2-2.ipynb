```python
def task1(ids):
    unique_tags = set()

    for user_tags in ids.values():
        unique_tags.update(user_tags)

    print(unique_tags)
```

```python
ids = {'user1': [213, 213, 213, 15, 213],
       'user2': [54, 54, 119, 119, 119],
       'user3': [213, 98, 98, 35]}
task1(ids)
```

```python
def task2(queries):
    word_count_distribution = {}

    for query in queries:
        word_count = len(query.split())
        if word_count not in word_count_distribution:
            word_count_distribution[word_count] = 0
        word_count_distribution[word_count] += 1

    total_queries = len(queries)

    for word_count in sorted(word_count_distribution.keys()):
        percentage = (word_count_distribution[word_count] / total_queries) * 100
        print(f"Поисковых запросов, содержащих {word_count} слов(а): {percentage:.2f}%")
```

```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]
task2(queries)
```
